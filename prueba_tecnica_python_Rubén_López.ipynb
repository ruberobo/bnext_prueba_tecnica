```python
import pandas as pd
import numpy as np
```

```python
usuarios=pd.read_excel('Mex Data Analyst Interview Case - Raw Data.xlsx')
usuarios
```

```python
dispositivos= pd.read_excel('Mex Data Analyst Interview Case - Raw Data.xlsx', 'users_devices' )
dispositivos
```

```python
tarjetas= pd.read_excel('Mex Data Analyst Interview Case - Raw Data.xlsx', 'cards' )
tarjetas
```

```python
transacciones= pd.read_excel('Mex Data Analyst Interview Case - Raw Data.xlsx', 'statements' )
transacciones
```

# pregunta 1

```python
#formato de fecha
usuarios['signup_date']=pd.to_datetime(usuarios['signup_date'])
```

```python
usuarios['mes'] = pd.DatetimeIndex(usuarios['signup_date']).month
```

```python
usuarios['hora']=pd.DatetimeIndex(usuarios['signup_date']).hour
```

```python
pregunta_1=usuarios[(usuarios['mes']==9)&(usuarios['hora']>=13)&(usuarios['hora']<=17)]
pregunta_1
```

```python
print('Registratos en septiembre entre las 13:00hrs y 17:00hrs =', len(pregunta_1))
```

# pregunta 2

```python
septiembre=usuarios[usuarios['mes']==9]
septiembre
```

```python
pregunta_2=pd.merge(septiembre, dispositivos,on='user_id')
pregunta_2
```

```python
respuesta_2=len(pregunta_2[pregunta_2['platform']=='IOS'])/len(pregunta_1)*100
```

```python
print('el porcentaje de ios de los registrados en septiembre es ',respuesta_2, '%')
```

# pregunta 5

```python
transacciones['operation_date']=pd.to_datetime(transacciones['operation_date'])
```

```python
transacciones['cuenta']=transacciones.sort_values(['operation_date'], ascending=[True]).groupby(['user_id']).cumcount() + 1
```

```python
transacciones
```

```python
pregunta_5=transacciones[transacciones['cuenta']==2]
```

```python
print('la media de la segunda compra es =',pregunta_5['amount'].mean())
```
